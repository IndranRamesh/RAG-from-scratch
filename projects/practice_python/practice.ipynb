````python
# Task 1 - Reverse String

# Version 1: Loop - Final Working logic

def reverse_string(s):
    rev = ''
    for char in s:
        rev = char + rev
    return rev

# tests
print(reverse_string("") == "")
print(reverse_string("a") == "a")
print(reverse_string("abc") == "cba")
print(reverse_string("hello") == "olleh")
````

````python
#  Task 2 - is_palindrome(with recursion + spaces/case handling)

def is_palindrome(text):
    def reverse_string(t):
        if len(t) <= 1:
            return t
        return reverse_string(t[1:]) + t[0]

    # Normalize : remove spaces and lowercase
    cleaned = text.replace(" ","").lower()
    reversed_cleaned = reverse_string(text).replace(" ","").lower()

    return cleaned == reversed_cleaned

print(is_palindrome("racecar") == True)
print(is_palindrome("hello") == False)
print(is_palindrome("A man a plan a canal Panama") == True)
print(is_palindrome("") == True)
````

````python
# Task 3 – Fibonacci (iterative + recursive)

def fib_iter(n):
    # iterative version
    temp_1, temp_2 = 0, 1
    value = 0
    for _ in range(n):
        temp_1, temp_2 = temp_2, temp_1 + temp_2
        value = temp_1
    return value


def fib_rec(n):
    # recursive version
    if n <= 1:
        return n
    value = fib_rec(n - 1) + fib_rec(n - 2)
    return value


# tests
print(fib_iter(0) == 0)
print(fib_iter(1) == 1)
print(fib_iter(6) == 8)
print(fib_iter(10) == 55)

print(fib_rec(0) == 0)
print(fib_rec(1) == 1)
print(fib_rec(6) == 8)
print(fib_rec(10) == 55)
````

````python
# Task 4 - Memoised Fibonacci + List helper

def fib_rec_memo(n,memo=None):
    if memo is None:
        memo = {} # return an empty dict if memo is empty
    if n in memo:
        return memo[n] # if  value n is in memo return the value of memo with the index of n
    if n <= 1:
        memo[n] = n
        return n
    value = fib_rec_memo(n - 1,memo) + fib_rec_memo(n-2 , memo)
    memo[n] = value
    return value

print(fib_rec_memo(0) == 0)
print(fib_rec_memo(n=1) == 15)
print(fib_rec_memo(n=2) == 1)
print(fib_rec_memo(n=6) == 10)
print(fib_rec_memo(10) == 55)
````

````python
# Part B – fib_list(n)
# Goal: Return a list of the first n Fibonacci numbers using your iterative version.

def fib_list(n):
    result = []
    for i in range(n):
        result.append(fib_iter(i))
    return result


# tests
print(fib_list(0) == [])
print(fib_list(1) == [0])
print(fib_list(5) == [0, 1, 1, 2, 3])
print(fib_list(7) == [0, 1, 1, 2, 3, 5, 8])
````
